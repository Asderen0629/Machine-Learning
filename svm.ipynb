```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from IPython.display import display
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.read_csv('heart.csv')
df.describe()
```

```python
x = df.loc[:,~df.columns.isin(['output'])]
y = df.loc[:,df.columns.isin(['output'])]
```

```python
train_x, test_x,train_y,test_y = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.corr().style.applymap(lambda x: 'color : red' if x >0.7 else 'color : white')
```

```python
model = svm.SVC(kernel= 'linear') #default rbf
model.fit(train_x,train_y)
print('---linear---')
display(confusion_matrix(model.predict(test_x),test_y))
model = svm.SVC(kernel= 'rbf') #default rbf
model.fit(train_x,train_y)
print('---rbf---')
display(confusion_matrix(model.predict(test_x),test_y))
model = svm.SVC(kernel= 'poly') #default rbf
model.fit(train_x,train_y)
print('---poly---')
display(confusion_matrix(model.predict(test_x),test_y))
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
LE = LabelEncoder()
temp = train['Activity'].copy()
train['Activity'] = LE.fit_transform(train['Activity'])
print(set(zip(temp,train['Activity'])))
temp = test['Activity'].copy()
test['Activity'] = LE.fit_transform(test['Activity'])
print(set(zip(temp,test['Activity'])))
```

```python
train_x = train.loc[:,~train.columns.isin(['Activity'])]
train_y = train.loc[:,train.columns.isin(['Activity'])]
test_x = train.loc[:,~test.columns.isin(['Activity'])]
test_y = train.loc[:,test.columns.isin(['Activity'])]
```

```python
model = svm.SVC(kernel='rbf')
model.fit(train_x,train_y)
display(confusion_matrix(model.predict(test_x),test_y))
print(model.score(test_x,test_y))
```

```python
model = svm.SVC(kernel='poly')
model.fit(train_x,train_y)
display(confusion_matrix(model.predict(test_x),test_y))
print(model.score(test_x,test_y))
```

```python
model = svm.SVC(kernel='linear')
model.fit(train_x,train_y)
display(confusion_matrix(model.predict(test_x),test_y))
print(model.score(test_x,test_y))
```

```python
df = pd.read_csv('log2.csv')
display(df)
display(df.describe())
display(df.corr().style.applymap(lambda x: 'color :red ' if x > 0.7 else  'color : white'))
```

```python
LE = LabelEncoder()
cols = ['Source Port', 'Destination Port', 'NAT Source Port', 'NAT Destination Port', 'Action']
for col in cols :
    temp = df[col]
    df[col] = LE.fit_transform(df[col])
    if col == 'Action' :
        print(set(zip(temp,df[col])))
```

```python
df['Action'].value_counts()
df2 = df.drop(columns=['Bytes','Packets','pkts_sent','pkts_received','Source Port','Destination Port','NAT Source Port','NAT Destination Port'])
```

```python
display(df2.corr().style.applymap(lambda x: 'color :red ' if x > 0.7 else  'color : white'))
```

```python
x = df2.loc[:,~df2.columns.isin(['Action'])]
y = df2.loc[:,df2.columns.isin(['Action'])]
```

```python
VIF = pd.DataFrame(columns=['VIF'],index = x.columns)
VIF['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
VIF.style.applymap(lambda x : 'color : yellow' if x >= 10 else 'color : white')
```

```python
train_x, test_x , train_y, test_y = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = svm.SVC()
model.fit(train_x,train_y)
confusion_matrix(model.predict(test_x),test_y)
```

```python
model = svm.SVC(kernel = 'linear')
model.fit(train_x,train_y)
confusion_matrix(model.predict(test_x),test_y)
```

```python
model = svm.SVC(kernel = 'poly')
model.fit(train_x,train_y)
confusion_matrix(model.predict(test_x),test_y)
```

```python
model = svm.SVC(kernel = 'poly',degree=2)
model.fit(train_x,train_y)
confusion_matrix(model.predict(test_x),test_y)
```
